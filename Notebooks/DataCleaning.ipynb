<center> Find columns that contain diagnosis age (or similar) and combine. Choose first non-null age.</center>

REF https://shecancode.io/filter-a-pandas-dataframe-by-a-partial-string-or-pattern-in-8-ways/#:~:text=Filter%20rows%20that%20match%20a%20given%20String%20in%20a%20column&text=We%20will%20use%20the%20Series,in%20the%20isin()%20function.

```python
mask = (df.columns.str.lower().str.contains("age") 
        & df.columns.str.lower().str.contains("diagnos")
        & ~df.columns.str.lower().str.contains("stage")
       )

# df.loc[:,mask].head()
cols = df.columns[mask]
# Not sure if bfill is the best choice here
df["Diagnosis Age"] = df.loc[:,cols].bfill(axis = 1).iloc[:,0]
df.drop(columns=cols).copy()
df.head()
```

<center>Create new Dataframe only with essential information</center>

```python
for col in df.columns:
    df.columns = df.columns.str.strip()
```

```python
note_cols = ["Gene", "Sample ID", "Cancer Type Detailed", "Mutation Type", "Variant Type", "HGVSc", "MS", 
        "Protein Change", "Functional Impact", "Diagnosis Age"]
rel_df = df[note_cols]
rel_df.head()
```

<center>Remove entries with no diagnosis age attached</center>

```python
removed_rows = rel_df.dropna(subset =["Diagnosis Age"], axis = 0)
removed_rows.head()
```

<center> Remove duplicate samples</center>

```python
print(removed_rows.shape)
```

```python
unique_samples = removed_rows["Sample ID"].unique()
print(unique_samples.shape)
```

```python
cleaned_df = removed_rows.drop_duplicates(subset=["Sample ID"], keep='first')
cleaned_df.shape
cleaned_df.info()
```

```python
cleaned_df.head()
```

```python
cleaned_df = cleaned_df.loc[cleaned_df["HGVSc"] == "ENST00000257430.4:c.835-8A>G"]
cleaned_df.shape
```

<center> Add column for early onset</center>

```python
cleaned_df = cleaned_df.copy()
cleaned_df["Early Onset"] = cleaned_df["Diagnosis Age"] <= 50
cleaned_df.head()
```

<center> Locate rows where Early Onset = True</center>

```python
early_onset = cleaned_df.copy()
early_onset = early_onset.loc[early_onset["Early Onset"] == True]
early_onset
```
